```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Sample dataset
data = {
    "text": [
        "Government announces new policy for students",
        "Aliens landed in India last night shocking news",
        "New AI technology improves healthcare",
        "Cure for all diseases discovered secretly",
        "Stock market shows positive growth today",
        "Scientists confirm flat earth theory"
    ],
    "label": [0, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# Split data
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.3)

# Convert text to numbers
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)

# Train model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Test prediction
sample = ["Breaking news: Scientists discover new planet"]
sample_vec = vectorizer.transform(sample)

prediction = model.predict(sample_vec)

print("Prediction (0=Real, 1=Fake):", prediction[0])
```
